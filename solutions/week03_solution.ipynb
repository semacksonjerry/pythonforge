```python
first = float(input("Enter first number: "))

print(first)
second = float(input("Enter second number: "))
print(second)


# Addition
sum = first + second

# Subtraction
sub = first - second

# Multi
prod = first * second

# Quotient
div = first / second

# reminder
mod = first % second

print(f" Sum: {sum} \n Product: {prod} \n Subtraction: {sub} \n Division: {div} \n Module: {mod}")
```
