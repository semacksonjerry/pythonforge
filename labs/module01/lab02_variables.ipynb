# Object Counters

A counter is an integer variable that allows you to count objects.

Counters typically have an initial value of zero, which increments to reflect the number of times a given object appears.

```python
str_counter = 0

for item in ("Alice", 30, "Programmer", None, True, "Department C"):
    if isinstance(item, str):
        str_counter += 1
```

```python
str_counter
```

# Note:

The line in the above example uses the expression str_counter += 1

 which is a shortcut for str_counter = str_counter + 1. 
 
 The += operator is known as the augmented addition operator.

# Accumulators

Accumulators are another common type of variable used in programming. 

An accumulator is a variable that you use to add consecutive values to form a total that you can use as an intermediate step in different calculations.

```python
numbers = [1, 2, 3, 4]
total = 0

for number in numbers:
    total =total + number
```

```python
total
```

calculate the mean of a list of numbers:

```python
total / len(numbers)
```

# Boolean Flags

Boolean flags help you manage control flow and decision-making in your programs. 

As their name suggests, these variables can be either True or False. 

You can use them in conditionals, while loops, and Boolean expressions.

```python
toggle = True

for _ in range(4):
    if toggle:
        print(f"✅ toggle is {toggle}")
        print("Do something...")
    else:
        print(f"❌ toggle is {toggle}")
        print("Do something else...")
    toggle = not toggle
```

# Loop Variables

```python
colors = [
    "red",
    "yellow",
    "green",
    "blue",
    "indigo",
    "violet"
]

for color in colors:
    print(color)
```

```python
for index, color in enumerate(colors):
    print(index, color)
```

# Data Storage Variables

Data storage variables allow you to work with containers of values, such as lists, tuples, dictionaries, or sets. 

For example, say that you’re writing a contact book application that uses a list of tuples to store the information of your contacts:

```python
contacts = [
    ("Rose", "111-2222-3333", "rose@example.com"),
    ("Esther", "111-2222-3333", "esther@example.com"),
    ("Lara", "111-2222-3333", "lara@example.com"),
    ("David", "111-2222-3333", "david@example.com"),
    ("Professor", "111-2222-3333", "prof@example.com"),
]
```

```python
for contact in contacts:
    print(contact)
```

```python
for index , contacts in enumerate(contacts):
    print(index, contact)
```
