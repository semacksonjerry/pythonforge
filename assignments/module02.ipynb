# Question3 

```python
num1 = int(input("Enter fisrt number: "))
num2 = int(input("Enter second number: "))

sum_result = num1 + num2
dif_result = num1 - num2
prod_result = num1 * num2
quo_result = num1 / num2
mud_result = num1 % num2
exp_result = num1 ** num2

print(f"Sum: {sum_result}")
print(f"Difference: {dif_result}")
print(f"Product: {prod_result}")
print(f"Quotient: {quo_result}")
print(f"Modulus: {mud_result}")
print(f"Exponent: {exp_result}")
```
