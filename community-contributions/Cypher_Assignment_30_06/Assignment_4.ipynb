Question 2
Ask the user to enter their full name.

Display:

Total number of characters
First 3 characters
Last 3 characters
Name in uppercase
Name in lowercase
Name in title case

```python
full_name = input("Enter your full name: ")
print(f"Total number of characters: {len(full_name)}")
print(f"first 3 characters: {full_name[:3]}")
print(f"last 3 characters: {full_name[-3:]}")
print(f"Name in uppercase: {full_name.upper()}")
print(f"Name in lowercase: {full_name.lower()}")
print(f"Name in title case: {full_name.title()}")
```

Question 7
Create a simple student information program.

Ask the user for:

Full Name
Age
Course
Score

Display all the information.

If the score is 50 or above, display:

Status : PASS

Otherwise display:

Status : FAIL

```python
student_name = input("Enter the student's name: ")
student_age = int(input("Enter the student's age: "))
course = input("Enter the course name: ")
score = float(input("Enter the student's score: "))
print("\n---------- Student Information ----------")
print(f"Full Name:          {student_name}")
print(f"Age:                {student_age}")
print(f"Course:             {course}")
print(f"Score:              {score}")
if score >= 50:
    print("Status: PASS")
else:
    print("Status: FAIL")
```

Question 8
Create a program that asks the user to enter three words.

Display:

Each word in uppercase
Each word in lowercase
The length of each word
The first and last letter of each word

Then ask the user to enter a number.

If the number is even display:

The number is even.

Otherwise display:

The number is odd.

```python
word1 = input("Enter the first word: ")
word2 = input("Enter the second word: ")
word3 = input("Enter the third word: ")
words = [word1, word2, word3]
for index, w in enumerate(words, start=1):
    print(f"uppercase: {w.upper()}")
    print(f"lowercase: {w.lower()}")
    print(f"length: {len(w)}")
    print(f"first letter: {w[0]}")
    print(f"last letter: {w[-1]}")

number = int(input("Enter a number: "))
if number % 2 == 0:
    print(f"The number is even.")
else:
    print(f"The number is odd.")
```
